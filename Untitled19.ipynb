```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
data = pd.DataFrame({
    'age': np.random.normal(30, 10, 100),
    'salary': np.random.normal(50000, 15000, 100),
    'experience': np.random.randint(1, 20, 100)
})
```

```python
plt.hist(data['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data['age'], kde=True, color='coral')
plt.title('Seaborn Histogram of Age')
plt.show()
```

```python
plt.boxplot(data['salary'])
plt.title('Boxplot of Salary')
plt.ylabel('Salary')
plt.show()
```

```python
sns.boxplot(y=data['salary'], color='lightgreen')
plt.title('Seaborn Boxplot of Salary')
plt.show()
```

```python
plt.scatter(data['age'], data['salary'], color='purple')
plt.title('Scatterplot: Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
sns.scatterplot(x='age', y='salary', data=data, hue='experience', palette='viridis')
plt.title('Seaborn Scatterplot: Age vs Salary (Colored by Experience)')
plt.show()
```
