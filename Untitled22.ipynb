```python
import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023],
    'Sales': [100, 150, 200, 250]
})

fig = px.line(df, x='Year', y='Sales', title='Sales Over Years')
fig.show()
```

```python
fig = px.bar(df, x='Year', y='Sales', title='Annual Sales')
fig.show()
```

```python
df2 = px.data.iris()
fig = px.scatter(df2, x="sepal_width", y="sepal_length",
                 color="species", size="petal_length", title="Iris Dataset Scatter")
fig.show()
```

```python
df3 = px.data.tips()
fig = px.pie(df3, values='tip', names='day', title='Tip Distribution by Day')
fig.show()
```

```python
fig = px.box(df3, x="day", y="total_bill", color="sex", title="Boxplot of Total Bill by Day and Gender")
fig.show()
```

```python
import numpy as np

# Time-series data
df_ts = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=100),
    'Value': np.random.randn(100).cumsum()
})

fig = px.line(df_ts, x='Date', y='Value', title='Random Time-Series')
fig.show()
```
